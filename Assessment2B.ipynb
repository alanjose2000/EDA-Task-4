###### Department wise codekata performence (pie chart)

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.cm as cm
from matplotlib.gridspec import GridSpec
import matplotlib.ticker as tk
import numpy as np

plt.figure(1, figsize=(25,15))
grid = GridSpec(2,2)
cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0,1,8)]
department = data.groupby('Department').agg('count')
```

```python
colors = [ "#FF5733", "#16A085", "#8E44AD"]
explode = (0.1, 0, 0) 
plt.pie(department["CodeKata Score"], labels=department.index, explode=explode, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Department wise Codekata Performance ")
plt.show()
```
